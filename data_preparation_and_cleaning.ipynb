```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/ml_projects/irrigation_dataset/dataset.csv', index_col=['date'], sep=',',parse_dates=['date'])
```

```python
data.head()
```

```python
(data['reference_evapotranspiration']).describe()
```

```python
data = data[(data['precipitation'] <= 60)]
data = data[(data['minimum_temperature'] >= 0)]
data = data[(data['minimum_relative_humidity'] >= 20)]
data = data[(data['total_solar_radiation'] <= 4000)]
```

```python
data = data.resample(rule='D').mean()
```

```python
data.head()
```

```python
X = data[[c for c in data.columns if c != 'reference_evapotranspiration']]
y = data[['reference_evapotranspiration']]
```

```python
X_train, X_test = X['2017-01-01':'2017-09-24'], X['2017-09-24':'2017-11-29']
y_train, y_test = y['2017-01-01':'2017-09-24'], y['2017-09-24':'2017-11-29']
```

```python
X_train.to_csv('/content/drive/MyDrive/ml_projects/irrigation_dataset/data/X_train_complete.csv')
y_train.to_csv('/content/drive/MyDrive/ml_projects/irrigation_dataset/data/y_train_complete.csv')
X_test.to_csv('/content/drive/MyDrive/ml_projects/irrigation_dataset/data/X_test.csv')
y_test.to_csv('/content/drive/MyDrive/ml_projects/irrigation_dataset/data/y_test.csv')
```
